```python
housing_df = pd.read_csv(housing_data_file, header=None, sep="\s+")
housing_df.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
housing_df
```

## Entendiendo los datos

### Descripción de las variables

1. `CRIM`:     Taza de criminalidad en la zona.
2. `ZN`:       Proporción de terreno recidencial para lotes mayores a 25k pies cuadrados
3. `INDUS`:    Proporción de industrias por ciudad
4. `CHAS`:     Variable dummy del río Charles (= 1 colimnda con el río; 0 lo contrario)
5. `NOX`:      Concentración de óxido nítrico (partes por 10 millones)
6. `RM`:       Promedio de habitaciones por vivienda
7. `AGE`:      Proporcion de unidades ocupadas por sus propietarios construidas antes de 1940
8. `DIS`:      Distancia ponderada a 5 centros de empleo de Boston
9. `RAD`:      Indice de accecibilidad a carreteras radiales
10. `TAX`:     Tasa de impuesto a la propiedad sobre el valor total por cada $10,000
11. `PTRATIO`: Relación alumno-profesor por ciudad
12. `B`:       1000(Bk - 0.63)^2 Bk es la proporción de personas de color por ciudad
13. `LSTAT`:   % Estatus inferior de la población
14. `MEDV`:    Valor medio de las viviendas ocupadas por sus propietarios en $1,000's

Buscamos si existe alguna correlación entre las variables y la variable objetivo (`MEDV`).

```python
cols = ["DIS", "INDUS", "CRIM", "RM", "MEDV"]
sns.pairplot(housing_df[cols], height=2.5)
plt.show()
```

Corroboramos las correlaciones mediante la matríz de coeficientes de correlación.

```python
cm = np.corrcoef(housing_df[cols].values.T)
sns.heatmap(cm,
  cbar=True,
  annot=True,
  yticklabels=cols,
  xticklabels=cols
)
plt.show()
```

Variables mejor correlacionadas con `MEDV`:
- `RM`: Alta correlación positiva, a mayor número de habitaciones mayor es la mediana del precio de las viviendas.
- `INDUS`: Baja correlación negativa, menor cantidad de industrias en la zona, menor es la mediana del precio de las viviendas.

## Entrenando el modelo

```python
from sklearn.model_selection import train_test_split

# Variables objetivo y predictora
x = housing_df["RM"].values.reshape(-1,1)
y = housing_df["MEDV"].values.reshape(-1,1)

# Creamos conjuntos de datos para entrenamiento y pruebas
X_train, X_test, Y_train, Y_test = train_test_split(x,y, random_state=42)

# Estandatizamos los datos
x_scaler = StandardScaler()
y_scaler = StandardScaler()

# Ajuste y transformación de las variables para entrenamiento
X_train_std = x_scaler.fit_transform(X_train)
Y_train_std = y_scaler.fit_transform(Y_train)

# Transformación de las variables de prueba
X_test_std = x_scaler.transform(X_test)
Y_test_std = y_scaler.transform(Y_test)

# Creamos el modelo
slr = LinearRegression()

slr.fit(X_train_std, Y_train_std)
```

## Análisis de resultados

```python
plt.scatter(X_train_std, Y_train_std, alpha=.5)
plt.plot(X_test_std, slr.predict(X_test_std), color="r")
plt.xlabel("Promedio de habitaciones [RM]")
plt.ylabel("Mediana del precio de las viviendas [MEDV]")
plt.show()
```

## Generando predicciones con el modelo

```python
rooms = 5
rooms_std = x_scaler.transform(np.array([rooms]).reshape(-1, 1))

predict = y_scaler.inverse_transform(slr.predict(rooms_std))
print(f"El precio de una casa de {rooms} habitaciones en Boston es de {predict}")
```

```python
plt.scatter(
  housing_df["RM"],
  housing_df["MEDV"]
)

plt.plot(rooms, predict, "ro")

plt.annotate(
  "Predicción",
  xy=(rooms-.6, predict+.6)
)

plt.title("Precio de una vivienda con 5 habitaciones")
plt.xlabel("Promedio de habitaciones [RM]")
plt.ylabel("Mediana del precio de las viviendas [MEDV]")
plt.show()
```

## Evaluando al modelo con $R^2$ y $MSE$

```python
from sklearn.metrics import mean_squared_error, r2_score

predicted_values = slr.predict(X_test_std)

trans_pred = y_scaler.inverse_transform(predicted_values)

mse = mean_squared_error(Y_test, trans_pred)
rmse = np.sqrt(mse)
r2 = r2_score(Y_test, trans_pred)

print("Evalucaicón del modelo con Mean Square Error: %.4f" % mse)
# Determinar el número de unidades que se desplaza el la predicción en relación al valor real
print("Evalucaicón del modelo con Root Mean Square Error: %.4f" % rmse)
print("Evalucaicón del modelo R^2: %.4f" % r2)
```

Al evaluar al modelo obtenemos
- MSE de 42.7496
- RMSE de 6.5383
- R2 de 0.3895


> 📌Los valores de MSE y RMSE se calcularon en base a la escala real de los valores de `MEDV`, por lo tanto su escala correspondonde a la de los valores reales.


El valor del error cuadratico medio (MSE) es relativamente alto (42.75), lo nos indica que el modelo no predice con presición. <br>
Mientras que RMSE con un valor de 6.5383 sugiere que las predicciones del modelo pueden desviarse considerablemente respecto de `MEDV`. <br>
Por otro lado al evaluar el modelo con $R^2$ obtenemos un valor 0.3895, lo que indica que la variable `RM` solo explica el 38.95% de la variabilidad de `MEDV.


En conclusión éste modelo deberá de entrenarse con más variables, pues el numéro de habitaciones no es el único factor que determina el precio de una vivienda.

### Gráfico de residuos

🎯 El objetivo de este gráfico es observar:
- Que los datos estén distribuidos simétricamente.
- Que los datos estén agrupados (acumulados) al rededor de los valores más pequeños del eje y (0, 0.5, 1 o ,1.5)
- Que no sigan un patron o tendencia aparente.

> 📌 Residuo = valor real - valor predicho, se estandarizan con el mismo método utilizado con los datos de entrenamiento. 

```python
plt.figure(figsize=(12,10))

trans_predictet_values = y_scaler.inverse_transform(predicted_values)
residuals = np.subtract(Y_test_std, predicted_values)

plt.subplot(2, 2, 1)
plt.scatter(
  trans_predictet_values,
  Y_test,
  alpha=.5
)

plt.xlabel("Valores predecidos para MEDV")
plt.ylabel("Valores actuales de MEDV")
plt.title("Predicción vs Reales")

plt.subplot(2, 2, 2)
plt.scatter(
  trans_predictet_values,
  residuals,
  alpha=.5
)

plt.ylabel("Residuos Estandarizados")
plt.xlabel("Valores predichos")
plt.title("Valores residuales")

plt.axhline(
  y=0,
  color="black",
  linestyle="dashed"
)
plt.show()
```

- El gráfico de la izquierda es la relación entre los valores reales (eje y) con los valores predichos por el modelo (eje x).
- El gráfico de la derecha vemos la relación de los valores predichos en su rango normal (eje x) con los residuos estandarizados.=
